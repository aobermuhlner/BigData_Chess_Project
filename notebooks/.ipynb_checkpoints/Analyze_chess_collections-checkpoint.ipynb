```python
import pandas as pd
import tqdm

def process_games(file_path, start_game=0, end_game=None, batch_size=500000):
    """
    Process games from a CSV file and expand the move sequences into individual rows,
    keeping a running sequence of all moves up to the current move. Adds a 'batch' column to indicate the batch number.
    Increments the count for duplicate positions across batches.

    :param file_path: Path to the CSV file.
    :param start_game: Index of the first game to process (0-indexed).
    :param end_game: Index of the last game to process (exclusive, None processes all).
    :param batch_size: The number of games per batch.
    """
    games = pd.read_csv(file_path)

    # Slice the DataFrame to only include the specified range of games
    games_to_process = games.iloc[start_game:end_game]
    
    # Initialize an empty DataFrame to store the expanded move data
    expanded_moves_df = pd.DataFrame(columns=['position', 'move_number', 'batch', 'count'])

    for index, game in tqdm.tqdm(games_to_process.iterrows(), total=games_to_process.shape[0]):
        # Check if 'moves' is not empty
        if not game['moves'] or pd.isna(game['moves']):
            continue  # Skip if moves are empty

        moves = eval(game['moves'])
        move_sequence = ''
        for move_number, move in enumerate(moves, start=1):
            move_sequence += move + ' '
            position = move_sequence.strip()

            # Check if the position already exists in the DataFrame
            if position in expanded_moves_df['position'].values:
                # Increment the count for the existing position
                expanded_moves_df.loc[expanded_moves_df['position'] == position, 'count'] += 1
            else:
                # Add a new row for the new position
                expanded_moves_df = expanded_moves_df.append({
                    'position': position,
                    'move_number': move_number,
                    'batch': (index - start_game) // batch_size + 1,  # Compute the batch number
                    'count': 1
                }, ignore_index=True)
    
    return expanded_moves_df

```

```python


# Example usage
file_path = '../../testData/splitted_games/processed_moves2.csv'
processed_games_df = process_games(file_path, start_game=0, end_game=10000, batch_size=2000)
print(processed_games_df.head())
```

```python
# Aggregate and count where move_number is 1, considering the batch
move_one_counts = processed_games_df[processed_games_df['move_number'] == 1].groupby(['batch', 'position']).size().reset_index(name='count')

print(move_one_counts)
```

```python
# Aggregate and count occurrences for each unique position sequence at each move number
move_counts = processed_games_df.groupby(['move_number', 'batch', 'position']).size().reset_index(name='count')

print(move_counts)
```

```python
import matplotlib.pyplot as plt

# Count unique positions per move sequence
unique_positions_per_move = processed_games_df.groupby('move_number')['position'].nunique()

# Get the total count of positions per move
total_positions_per_move = processed_games_df.groupby('move_number').size()

# Plot the data
plt.figure(figsize=(10, 6))

# Bar plot for unique positions
unique_positions_per_move.plot(kind='bar', alpha=0.5, label='Unique Positions')

# Line plot for total count of positions
total_positions_per_move.plot(kind='line', color='red', marker='o', label='Total Count of Positions')

plt.title('Unique Positions and Total Count of Positions per Move Sequence')
plt.xlabel('Move Number')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Keep the x-axis labels readable
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Count unique positions per move sequence
unique_positions_per_move = processed_games_df.groupby('move_number')['position'].nunique()

# Get the total count of positions per move
total_positions_per_move = processed_games_df.groupby('move_number').size()

# Calculate the percentage of unique positions and duplicates per move
percentage_unique = (unique_positions_per_move / total_positions_per_move) * 100
percentage_duplicates = 100 - percentage_unique

# Create a DataFrame for plotting
percentage_df = pd.DataFrame({
    'Unique Positions': percentage_unique,
    'Duplicate Positions': percentage_duplicates
})
```

```python
import matplotlib.pyplot as plt

# Create the plot
plt.figure(figsize=(12, 6))
ax = percentage_df.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], figsize=(12, 6))
plt.title('Percentage of Unique and Duplicate Positions per Move Sequence')
plt.xlabel('Move Number')
plt.ylabel('Percentage')
plt.xticks(rotation=0)  # Keep the x-axis labels readable
plt.legend(loc='upper right')

# Annotate bars with the percentage of duplicate positions
for i, rect in enumerate(ax.patches[len(ax.patches)//2:]):  # Iterate over the duplicate position bars
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2.0, 
            rect.get_y() + height / 2.0, 
            f'{percentage_duplicates.iloc[i]:.1f}%', 
            ha='center', 
            va='center', 
            color='black',
            fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by move_number and batch, then count unique positions
unique_positions_per_move_batch = processed_games_df.groupby(['move_number', 'batch'])['position'].nunique().reset_index(name='unique_count')
unique_positions_per_move_batch = unique_positions_per_move_batch[unique_positions_per_move_batch['move_number'] <= 10]
# Group by move_number and batch, then count total positions
total_positions_per_move_batch = processed_games_df.groupby(['move_number', 'batch']).size().reset_index(name='total_count')

# Merge the unique and total counts
merged_df = pd.merge(unique_positions_per_move_batch, total_positions_per_move_batch, on=['move_number', 'batch'])

# Calculate the percentage of unique and duplicate positions for each batch within each move number
merged_df['percentage_unique'] = (merged_df['unique_count'] / merged_df['total_count']) * 100
merged_df['percentage_duplicates'] = 100 - merged_df['percentage_unique']

# Pivot the data for plotting
percentage_unique_pivot = merged_df.pivot(index='move_number', columns='batch', values='percentage_unique')
percentage_duplicates_pivot = merged_df.pivot(index='move_number', columns='batch', values='percentage_duplicates')

# Plotting
plt.figure(figsize=(12, 6))
percentage_unique_pivot.plot(kind='bar', stacked=True, ax=plt.gca(), colormap='cool')
percentage_duplicates_pivot.plot(kind='bar', stacked=True, ax=plt.gca(), colormap='autumn', alpha=0.6)

plt.title('Percentage of Unique and Duplicate Positions per Move Sequence by Batch')
plt.xlabel('Move Number')
plt.ylabel('Percentage')
plt.legend(title='Batch', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
